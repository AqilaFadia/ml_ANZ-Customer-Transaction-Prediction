Train-Test-Split

```python
from sklearn.model_selection import train_test_split
 
X = anz.drop(["amount"],axis =1)
y = anz["amount"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 123)
```

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')
```

The division of the Train-Test-Split dataseet 80:20

Standardization 

```python
from sklearn.preprocessing import StandardScaler
 
numerical_features = ['balance', 'age']
scaler = StandardScaler()
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

```python
from sklearn.preprocessing import StandardScaler
 
numerical_features = ['balance', 'age']
scaler = StandardScaler()
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

The following data frame for the analysis of the three models

```python
# Prepare a dataframe for model analysis
models = pd.DataFrame(index=['train_mse', 'test_mse'], 
                      columns=['KNN', 'RandomForest', 'Boosting'])
```

Model Development with K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
 
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train, y_train)
 
models.loc['train_mse','knn'] = mean_squared_error(y_pred = knn.predict(X_train), y_true=y_train)
```

Model Development with Random Forest

```python
# Import the required libraries
from sklearn.ensemble import RandomForestRegressor
 
# create predictive models
RF = RandomForestRegressor(n_estimators=50, max_depth=16, random_state=55, n_jobs=-1)
RF.fit(X_train, y_train)
 
models.loc['train_mse','RandomForest'] = mean_squared_error(y_pred=RF.predict(X_train), y_true=y_train)          
```

Model Development with Boosting Algorithm

```python
from sklearn.ensemble import AdaBoostRegressor
 
boosting = AdaBoostRegressor(learning_rate=0.05, random_state=55)                             
boosting.fit(X_train, y_train)
models.loc['train_mse','Boosting'] = mean_squared_error(y_pred=boosting.predict(X_train), y_true=y_train)
```

MODEL EVALUATION

```python
# Perform scaling of the numeric features in X_test so that it has mean = 0 and variance = 1
X_test.loc[:, numerical_features] = scaler.transform(X_test[numerical_features])
```

```python
mse = pd.DataFrame(columns=['train', 'test'], index=['KNN','RF','Boosting'])
 
# Create a dictionary for each algorithm used
model_dict = {'KNN': knn, 'RF': RF, 'Boosting': boosting}
 
# Calculate the Mean Squared Error of each algorithm on the train and test data
for name, model in model_dict.items():
    mse.loc[name, 'train'] = mean_squared_error(y_true=y_train, y_pred=model.predict(X_train))/1e3 
    mse.loc[name, 'test'] = mean_squared_error(y_true=y_test, y_pred=model.predict(X_test))/1e3
 
# Call mse
mse
```

```python
fig, ax = plt.subplots()
mse.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)
```

```python
prediksi = X_test.iloc[:1].copy()
pred_dict = {'y_true':y_test[:1]}
for name, model in model_dict.items():
    pred_dict['prediksi_'+name] = model.predict(prediksi).round(1)
 
pd.DataFrame(pred_dict)
```
